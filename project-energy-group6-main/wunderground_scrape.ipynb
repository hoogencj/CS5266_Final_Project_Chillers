```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.firefox.firefox_binary import FirefoxBinary
from selenium.webdriver.support import expected_conditions as EC

import datetime
from datetime import date
from datetime import timedelta
```

```python
import pandas as pd

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',options=chrome_options)

start = date(2020, 1, 1)
end = date(2021, 1, 1)
day = timedelta(days=1)

def date_time(time, start):
  dstr = start.strftime('%Y/%m/%d') + ' ' + time
  return datetime.datetime.strptime(dstr, '%Y/%m/%d %I:%M %p')


dfs = pd.DataFrame()
while start <= end:
  url = 'https://www.wunderground.com/history/daily/us/tn/nashville/date/' + start.strftime('%Y-%m-%d')
  wd.get(url)

  try:
    tables = WebDriverWait(wd,20).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "table")))

    if len(tables) > 1:
      df = pd.read_html(tables[1].get_attribute('outerHTML'))
    else:
      print(tables[0])
      df = False

    if df: 
      df = df[0].dropna()
      df['Date'] = df['Time'].apply(lambda time: date_time(time, start))
      dfs = dfs.append(df)
      print(df.head())
    
  except:
    continue
    
  finally:
    start += day
```

```python
dfs.to_csv('nashville-weather.csv')
```
